## ***5. Hypothesis Testing***

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Answer Here.Null Hypothesis ($H_0$): All countries spend the same average amount. (i.e., There is no significant difference in Total_Price across different countries).Alternate Hypothesis ($H_1$): At least one country spends a different average amount compared to the others. (i.e., Different countries spend different amounts).

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value
# from Visulaization-Relationship between Aggregate rating and votes

from scipy import stats
# H0= No correaltion between Aggregate rating and votes
# H1 = Signifiant correlation between Aggregate rating and votes
#performing pearson correlation  test

corr,p_value = stats.pearsonr(df['Aggregate rating'],df['Votes'])
alpha = 0.05
print("Pearson Coorelation Test Result")
print(f"Correaltion   :{corr:.4f}")
print(f"p_value      :{p_value:.4f}")
if p_value < alpha :
    print("\n Accept Alternate Hypothesis (H1)")
    print("Signifiant correlation between Aggregate rating and votes")
else:
    print("\n Accept Null Hypothesis (H0)")
    print(" No correlation between Aggregate rating and votes")
````

### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
# Handling Outliers & Outlier treatments
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv(r"Zomato Merged Dataset.csv")
print(f" Dataset Shape : {df.shape}")
print("="*55)
numeric_cols=['Price range','Aggregate rating','Votes','Average Cost for two']
# detetcing outliers
fig,axes=plt.subplots(1,len(numeric_cols),figsize=(20,5))
fig.suptitle("Boxplot -outliers detection defore treatment", fontsize=14)
for i , col in enumerate(numeric_cols):
    axes[i].boxplot(df[col].dropna())
    axes[i].set_title(col)
    axes[i].set_xlabel(col)
plt.tight_layout()
plt.show()
print("="*55)
# detecting outliers using IQR method
print("IQR method for outliers detection")
outlier_report=[]
for col in numeric_cols:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR = Q3-Q1

    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR

    outlier=df[(df[col]< lower)|(df[col]>upper)]
    outlier_report.append({
        "Column"       : col,
        "Q1"           : round(Q1,2),
        "Q3"           : round(Q3,2),
        "IQR"          : round(IQR,2),
        "Lower Bound"  : round(lower,2),
        "Upper Bound"  : round(upper,2),
        "Outlier Count": len(outlier),
         "Outlier %"   : round(len(outlier) / len(df) * 100, 2)
    })

outlier_df=pd.DataFrame(outlier_report).set_index("Column")
print(outlier_df.to_string())
print("="*55)

# Outlier Treatment method
print("Outlier Treatment Method")
df_treated=df.copy()
# Treatment-1 Capping / Winsorization 
print("Treatment Capping (Winsorization)")
cap_cols=['Votes','Average Cost for two']
for col in cap_cols:
    Q1= df_treated[col].quantile(0.25)
    Q3= df_treated[col].quantile(0.75)
    IQR= Q3-Q1

    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR

    df_treated[col]=df_treated[col].clip(lower=lower,upper=upper)
    after = df_treated[col].describe()
    print(f"   {col:<15} capped [{lower:,.0f} to {upper:,.0f}]")
#Visualize before and after Treatment
print("Before Vs After Treatment")
print("="*55) 

fig,axes = plt.subplots(1,len(cap_cols), figsize=(15,10))
fig.suptitle("After Outlier Treatment", fontsize =14)
for i , col in enumerate(cap_cols):   
    axes[i].boxplot(df_treated[col].dropna())
    axes[i].set_title(f"{col}AFTER")
    axes[i].set_xlabel("After Capping")

print(" Treated Outliers")
df_treated.to_csv("Zomato_treated outliers.csv")
print("Saved")
````

### 3. Categorical Encoding

````python
# Encode your categorical columns
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv("Zomato_treated outliers.csv")
cat_cols = ['Restaurant Name',
            'City','Address',
            'Locality','Locality Verbose',
            'Cuisines,Currency','Has Table booking',
            'Has Online delivery','Is delivering now', 
            'Switch to order menu','Rating color',
            'Rating text','Country']
print(f"Categorical Columns  : {len(cat_cols)}")
drop_cols = ['Restaurant Name','Restaurant ID','Address','Locality Verbose','Switch to order menu']
df.drop(columns=drop_cols,inplace=True,errors='ignore')
# performing One Hot Encoding
one_hot_encode = ['Country','Country Code','Currency','Rating color','Rating text']
df = pd.get_dummies(df, columns = one_hot_encode, drop_first=True, dtype=int)
# label encoder
label_cols = ['Has Table booking','Has Online delivery','Is delivering now']
le = LabelEncoder()
for col in label_cols:
    if col in df.columns:
        df[col]=le.fit_transform(df[col])
# Frequency / target encoding
for col in ['City','Locality','Cuisines']:
    if col in df.columns:
        freq = df[col].value_counts()/len(df)
        df[col]=df[col].map(freq)

print(f"Encoded Dataset Shape :{df.shape}")
print(df.head())
df.to_csv("Zomato Encoded Dataset.csv",index=False)
print("saved")
````

Dataset shape is (9551, 63) and every single column has already been successfully converted into clean numbers (integers and floats). Since there zero raw text columns left, text NLP steps like Expand Contraction, Lower Casing, and Removing Punctuations have absolutely nothing to do.
